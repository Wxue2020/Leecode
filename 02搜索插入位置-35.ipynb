# 官方题解
在排序数组中寻找是否存在一个目标值，利用二分法在O(logn) 的时间内找到是否存在目标值
- 在一个有序数组中找第一个大于等于target 的下标：nums[pos−1]<target≤nums[pos]
- 存在一种情况是 target 大于数组中的所有数，此时需要插入到数组长度的位置

复杂度
- 时间复杂度：O(logn)，其中 nn 为数组的长度。二分查找所需的时间复杂度为 O(logn)，每次循环都会丢弃一半的查找空间
- 空间复杂度：O(1)。我们只需要常数空间存放若干变量，使用一个整数型变量mid来记录中间值



### 二分查找算法
【待查找数是整数，知道范围，可以使用二分查找】
#### 算法：
- 先设定左侧下标left和右侧下标right，再计算中间下标mid
- 根据nums[mid]和target之间的大小进行判断，相等则直接返回下标，nums[mid]<target则left右移，num[mid]>target则right左移
- 查找结束没有相等值则返回left，该值为插入位置

````python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        l, r = 0, len(nums)-1

        while l <= r:
            mid = (l+r)//2  # python中是向下取整
            if target == nums[mid]:
                return mid
            elif target < nums[mid]:
                r = mid - 1
            else:
                l = mid + 1
        return l
````
